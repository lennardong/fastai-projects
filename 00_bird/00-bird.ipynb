```python
## fastbook, fastai, fastcore
## fastai-projects/00-bird.ipynb
import fastbook as fb
import fastai.vision.all as fav
import fastai as fa
import fastcore as fc
import fastdownload as fd

import duckduckgo_search as ddg
```

```python
search = ddg.DDGS()
urls = [img["image"] for img in search.images('bird photos')]

urls[0]
```

```python
dest = './00_bird/bird.jpg'
fd.download_url(urls[0], dest)
im = fav.Image.open(dest)

im.to_thumb(512, 512)
```
